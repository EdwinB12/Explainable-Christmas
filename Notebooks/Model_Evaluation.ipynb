# Model Evaluation

#### This notebook will briefly evaluate the trained models performance. 


```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras
import seaborn as sns
import pandas as pd
```

## Load Data and Model

```python
# Load Model
model = keras.models.load_model('../outputs/model')

# Load Test Dataset
x_test = np.load('../outputs/x_test.npy')[:, ::2, ::2]
y_test = np.load('../outputs/y_test.npy')
```

```python
# Predict with Model
y_pred = model.predict(x_test)
```

## Metrics

```python
global_metrics = model.evaluate(x_test, y_test[:, ::2, ::2])
```

```python
# Generate Metrics for Each entry in the test dataset
metrics = np.asarray([model.evaluate(np.expand_dims(x, axis=0), np.expand_dims(y, axis=0)) for x, y in zip(x_test, y_test[:, ::2, ::2])])
```

```python
metric_df = pd.DataFrame(metrics, columns=[met.name for met in model.metrics])
```

```python
# Plot all metrics
fig, axes = plt.subplots(2, int(np.ceil(len(metric_df.columns)/2)), figsize=(20, 10), sharey=True)
for i, name in enumerate([met.name for met in model.metrics]):
    ax = axes.flat[i]
    sns.histplot(data=metric_df, x=name, ax=ax, bins=10, stat='probability')
    #ax.scatter(range(0, len(ordered_metrics)), ordered_metrics[:, i], s=2, c=range(0, len(ordered_metrics)))
    ax.set_title(name)
```

## Resize to make better looking pictures

```python
def resize_image(input_image, new_shape=[48, 48], **kwargs):
    """
    Resizes set of images. Note the dimension of the input must (num_of_examples, xdim, ydim, channels).
    """
    return tf.image.resize(tf.convert_to_tensor(input_image), new_shape, **kwargs).numpy()
```

```python
x_test = resize_image(x_test)
y_pred = resize_image(y_pred)
```

## Plotting Lots of Examples

```python
def plotting(x, y_true, y_pred, metrics, idx, save=False):
    fig, axes = plt.subplots(2, 3, figsize=(10, 6))
    im = axes[0, 0].imshow(x[idx, :, :, 0], cmap='gray')
    plt.colorbar(im, ax=axes[0, 0])
    im = axes[0, 1].imshow(x[idx, :, :, 1], cmap='bwr')
    plt.colorbar(im, ax=axes[0, 1])
    im = axes[0, 2].imshow(x[idx, :, :, 2], cmap='bwr')
    plt.colorbar(im, ax=axes[0, 2])
    im = axes[1, 0].imshow(x[idx, :, :, 3], cmap='bwr')
    plt.colorbar(im, ax=axes[1, 0])
    im = axes[1, 1].imshow(y_true[idx, :, :, 0], vmin=0, vmax=1)
    plt.colorbar(im, ax=axes[1, 1])
    im = axes[1, 2].imshow(y_pred[idx, :, :, 0], vmin=0, vmax=1)
    plt.colorbar(im, ax=axes[1, 2])
    for ax, title in zip(axes.flat, ['Seismic', 'Energy', 'Semblance', 'Edge Detection', 'True Model', 'Predicted Model']):
        ax.set_title(title)
        
    fig.suptitle(f'Index:{idx}: Loss:{metrics[idx, 0]:.3f}, Accuracy: {metrics[idx, 5]:.3f}')
    if save:
        fig.savefig(f'Figures/Prediction_Example_idx{idx}')
```

```python
loss_idx = metrics[:, 0].argsort()
```

#### Plot several examples, starting from lowest Loss to Highest Loss

```python
for idx in range(0, len(loss_idx), 75):
    plotting(x_test, y_test, y_pred, metrics, loss_idx[idx])
```
